# Task 1: User Login Check
**Objective:** Validate user credentials.
**Real-World Application:** Authentication systems.


```python
# ---------------------------------------------
# Function: login_check
# Purpose : Verify whether username and password match stored credentials
# Inputs  : username (str), password (str)
# Output  : Login status message (str)
# ---------------------------------------------
def login_check(username: str, password: str) -> str:
    """
    Validate user login credentials.
    Returns login status message.
    """
    correct_username = "admin"
    correct_password = "1234"

    if username == correct_username and password == correct_password:
        return "Login Successful"
    return "Invalid Credentials"
```

```python
# Test Credentials
print(login_check("admin", "1234"))
```

# Task 2: Pass / Fail Analyzer
**Objective:** Analyze student marks and count pass/fail.
**Real-World Application:** Academic evaluation systems.


```python
# ---------------------------------------------
# Function: analyze_results
# Purpose : Count total pass and fail students
# Input   : List of marks
# Output  : Dictionary containing summary
# ---------------------------------------------
def analyze_results(marks: list) -> dict:
    """
    Analyze student marks and return pass/fail summary.
    """
    # Count students scoring >= 50 (Pass condition)
    pass_count = sum(mark >= 50 for mark in marks)
    fail_count = len(marks) - pass_count

    return {
        "Total Students": len(marks),
        "Passed Students": pass_count,
        "Failed Students": fail_count
    }
```

```python
marks = [45, 78, 90, 33, 60]
result = analyze_results(marks)

print(result)
```

# Task 3: Simple Data Cleaner
**Objective:** Clean and standardize names.
**Real-World Application:** Data preprocessing before analysis.


```python
# ---------------------------------------------
# Function: clean_names
# Purpose : Remove extra spaces and convert names to lowercase
# Input   : List of names
# Output  : Cleaned list of names
# ---------------------------------------------
def clean_names(names: list) -> list:
    """
    Remove extra spaces and convert names to lowercase.
    """
    return [name.strip().lower() for name in names]
```

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned = clean_names(names)

print(cleaned)
```

# Task 4: Message Length Analyzer
**Objective:** Measure message length and flag long messages.
**Real-World Application:** Text validation systems.


```python
# ---------------------------------------------
# Function: analyze_messages
# Purpose : Compute length of each message and flag if > 10 characters
# Input   : List of messages
# Output  : List of dictionaries with analysis
# ---------------------------------------------
def analyze_messages(messages: list) -> list:
    """
    Analyze message lengths and flag messages longer than 10 characters.
    """
    return [
        {
            "Message": msg,
            "Length": len(msg),
            "Long_Message": len(msg) > 10
        }
        for msg in messages
    ]
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
analysis = analyze_messages(messages)

for item in analysis:
    print(item)
```

# Task 5: Error Message Detector
**Objective:** Count total ERROR entries in logs.
**Real-World Application:** Log monitoring systems.


```python
# ---------------------------------------------
# Function: count_errors
# Purpose : Count occurrences of the word "ERROR"
# Input   : List of log entries
# Output  : Integer count of ERROR logs
# ---------------------------------------------
def count_errors(logs: list) -> int:
    """
    Count total number of 'ERROR' entries in logs.
    """
    return logs.count("ERROR")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
print("Total ERROR Count:", count_errors(logs))
```
